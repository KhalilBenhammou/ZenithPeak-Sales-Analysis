### Data Cleaning & Wrangling

```python
df = df_sales.merge(
    df_customers, how="left", left_on="Customer Name Index", right_on="Customer Index"
)
```

```python
df = df.drop(columns=["Customer Index"])
df.head()
```

```python
df = df.merge(
    df_products, how="left", left_on="Product Description Index", right_on="Index"
).drop(columns=["Index"])
df.head()
```

```python
df = df.merge(
    df_regions, how="left", left_on="Delivery Region Index", right_on="id"
).drop(columns=["id"])
df.head()
```

```python
df = df.merge(
    df_state_regions[["State Code", "Region"]],
    how="left",
    left_on="state_code",
    right_on="State Code",
)
df.head()
```

```python
df = df.merge(df_budgets, how="left", on="Product Name")
```

```python
df.columns = df.columns.str.lower()
df.columns.values
```

```python
cols_to_keep = [
    "ordernumber",
    "orderdate",
    "customer names",
    "channel",
    "product name",
    "order quantity",
    "unit price",
    "line total",
    "total unit cost",
    "state_code",
    "county",
    "state",
    "region",
    "latitude",
    "longitude",
    "2017 budgets",
]
```

```python
df = df[cols_to_keep]
df.head()
```

```python
df = df.rename(
    columns={
        "ordernumber": "order_number",
        "orderdate": "order_date",
        "customer names": "customer_name",
        "product name": "product_name",
        "order quantity": "order_quantity",
        "unit price": "unit_price",
        "line total": "revenue",
        "total unit cost": "cost",
        "state_code": "state",
        "state": "state_name",
        "latitude": "lat",
        "longitude": "lon",
        "2017 budgets": "budget",
    }
)
df.head()
```

```python
df.loc[df["order_date"].dt.year != 2017, "budget"] = pd.NA

df[["order_date", "product_name", "revenue", "budget"]].head(5)
```
